```python
import pandas as pd
!pip install openpyxl

# Load the Excel file and select the "all" tab
file_path = "NYC_summons.xlsx"
df = pd.read_excel(file_path)
print(df.head())
```

```python
total_summons = df.loc[df['year'] == 'total'].iloc[0, 1:].values
unsafe_speeding = df.loc[df['year'] == 'unsafe speed'].iloc[0, 1:].values
unsafe_speed_norm = df.loc[df['year'] == 'unsafe speed'].iloc[0, 1:].values / df.loc[df['year'] == 'total'].iloc[0, 1:].values
```

```python
import matplotlib.pyplot as plt

years = range(2013, 2025)

# Plot total summons
plt.figure(figsize=(10, 6))
plt.bar(years, total_summons, color='#1f78b4', alpha=0.7)
plt.xlabel('Year')
plt.ylabel('Total Summons')
plt.title('Total Summons by Year')
plt.xticks(years, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plot unsafe speed summons
plt.figure(figsize=(10, 6))
plt.bar(years, unsafe_speeding, color='#1f78b4', alpha=0.7)
plt.xlabel('Year')
plt.ylabel('Unsafe Speed Summons')
plt.title('Unsafe Speed Summons by Year')
plt.xticks(years, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plot unsafe speed normalized
plt.figure(figsize=(10, 6))
plt.bar(years, unsafe_speed_norm, color='#1f78b4', alpha=0.7)
plt.xlabel('Year')
plt.ylabel('Unsafe Speed Normalized')
plt.title('Unsafe Speed Normalized by Year')
plt.xticks(years, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
